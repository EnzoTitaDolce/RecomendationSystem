```python
import pandas as pd
import ast
```

Importamos los archivos

```python
games = pd.read_csv('../dfSteamCleaned.csv')
user_items = pd.read_csv('../user_itemsCleaned.csv')
```

Eliminaremos los registros que no tengan información sobre el género

```python
games.dropna(subset='genres', inplace=True)
```

```python
games.info()
```

```python
games.head()
```

Aplicamos el método literal_eval de la librería ast para convertir los strings de la columna genres en listas reales

```python
type(games['genres'].iloc[0])
```

```python
games['genres'] = games['genres'].apply(ast.literal_eval)
```

```python
games = games.explode('genres')
```

```python
type(games['genres'].iloc[0])
```

Vemos que las listas ya se desanidaron

```python
games.head()
```

Comprobaremos los valores nulos en la columna genres y year. Los registros que no aporten información en estas columnas no nos sirven para esta función. Recordemos que en el archivo donde hicimos el ETL reemplazamos por 0 los valores nulos para el año de lanzamiento.

```python
games.isna().sum()
```

```python
games = games[(games['year']!=0)]
```

```python
games.info()
```

Del dataframe games nos quedaremos con las columnas genres, year y id.

```python
games = games[['id','genres','year']]
```

Del df user_items obtendremos el id de usuario, el id del juego y la cantidad de horas jugadas.
Consideraremos que las horas jugadas corresponden al año de lanzamiento.

```python
user_items.info()
```

```python
user_items = user_items[['user_id','item_id','playtime_forever']]
```

Analizamos si hay duplicados

```python
user_items.duplicated().sum()
```

```python
user_items[user_items.duplicated()]
```

```python
user_items.drop_duplicates(inplace=True)
```

Eliminamos los registros con valores nulos y cero en el campo playtime_forever.

```python
user_items.dropna(subset='playtime_forever', inplace=True)
user_items = user_items[(user_items['playtime_forever']!=0)]
```

```python
user_items.info()
```

```python
user_items['playtime_forever'].isna().sum()
```

Unimos los df haciendo coincidir el id del juego. Unimos el df con datos de los items del usuario y el df con datos del juego.

```python
df=pd.merge(user_items,games, left_on='item_id', right_on='id', how='inner')
```

```python
df.info()
```

Los valores del campo playtime_forever están en minutos así que los convertimos a hora dividiendolos en 60.

```python
df['playtime_forever'] = df['playtime_forever']/60
```

```python
df
```

Calculamos cuál sería el máximo tiempo de juego posible teniendo en cuenta el año de lanzamiento del juego y el 2021 qué es el año máximo que detectamos en el EDA. Si algún item supera este tiempo de juego, claramente, es un valor erroneo.

Para eso vamos a calcular la columna max_possible_playtime haciendo la diferencia entre 2023 y el año de lanzamiento y multiplicando ese valor por 8760 que representa la cantidad de horas en un año promedio (no bisiesto).

```python
df['max_possible_playtime'] = (2021-df['year'])*8760
```

Ahora veremos si hay algún registro que supere ese máximo tiempo de juego posible. Tengamos en cuenta que estamos considerando el supuesto extremo de que un jugador pueda pasar 24hs diarias jugando. Así que con toda certeza podemos afirmar que si algún registro del campo playtime_forever supera al valor de max_possible_playtime se trata de un error.

```python
excedidos = df[(df['playtime_forever']>df['max_possible_playtime'])]
excedidos
```

No hay ningún registro con valores de playtime_forever excedido. Podemos eliminar los dataframes que ya no necesitamos.

```python
del(excedidos, user_items, games)
```

```python
df.info()
```

Podemos eliminar también la columna item_id porque ya no es necesaria.

```python
df.drop(columns='item_id', inplace=True)
```

Agrupamos el dataframe según género y usuario y ordenamos según playtime_forever.

```python
df = df.groupby(['user_id','genres','year'])['playtime_forever'].sum().reset_index()
```

```python
df
```

Debemos recordar que la columna playtime_forever se refiere a la cantidad de horas que el usuario jugó en total y no solo durante el año de lanzamiento del juego. Es decir, si la columna playtime_forever muestra un valor de 5000 para el año 2007 quiere decir que el usuario jugó 5000 horas a un/unos juego/s lanzado/s en el año 2007, pero no necesariamente jugó todas esas horas en el año 2007. Por eso hay algunos valores superiores a 8784 horas que es la cantidad de horas máximas que puede haber en un año bisiesto.

```python
df[(df['playtime_forever']>=8784)]
```

Calculamos el usuario con más horas de juego para cada género.

```python
total_per_genre = df.groupby(['genres','user_id'])['playtime_forever'].sum().reset_index()
total_per_genre
```

```python
player_with_max_hours = total_per_genre.loc[total_per_genre.groupby('genres')['playtime_forever'].idxmax()]
player_with_max_hours
```

```python
del(total_per_genre)
```

Usaremos el dataframe player_with_max_hours para extraer el nombre de usuario de los usuarios con más horas acumuladas para cada género. Con esos nombres extraeremos la cantidad de horas por año del dataframe df.

```python
resultados = pd.merge(player_with_max_hours, df, left_on='user_id', right_on='user_id', how='inner')
resultados
```

```python
resultados.drop(columns='genres_y',inplace=True)
```

```python
resultados.info()
```

```python
resultados.head()
```

Guardamos los resultados en un archivo csv

```python
resultados.to_csv('userForGenreResults.csv')
```
