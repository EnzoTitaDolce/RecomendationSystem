```python
import pandas as pd
import numpy as np
```

Cargamos el dataframe analyzed_reviews que tiene las rese√±as ya calificadas y el dataframe dfSteamCleaned para obtener el nombre de los desarrolladores.

```python
analyzed_reviews = pd.read_csv('../user_reviews_df_cleaned.csv')
games = pd.read_csv('../dfSteamCleaned.csv')
```

Unimos los dataframes coincidiendo la columna item_id de analyzed_reviews con la columna id de games.

```python
df = pd.merge(analyzed_reviews, games, right_on='id', left_on='item_id' )
```

```python
df
```

Conservaremos solo las columnas necesarias para la funci√≥n.

```python
df = df[['item_id','sentiment_analysis','developer']]
```

```python
df
```

```python
len(df['developer'].unique())
```

```python
resultados = df.groupby('developer')['sentiment_analysis'].value_counts().unstack().fillna(0)
```

Exportamos el dataframe a un archivo csv

```python
df.to_csv('developerReviewsResults.csv', index=False)
```
