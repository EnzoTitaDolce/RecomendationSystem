Para esta función vamos a utilizar el archivo dfSteamCleaned para obtener el nombre del desarrollador, el precio de los juegos y el año de lanzamiento.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../dfSteamCleaned.csv')
```

```python
df
```

Conservaremos las columnas id, developer, price y year

```python
df = df[['id','developer','price','year']]
```

```python
df
```

Convertiremos los precios a tipo float. Para ellos usaremos la función conversor que convertirá los valores con formato numérico a float y al resto los reemplazara por cero.

```python
def conversor(x):
    try:
        x=float(x)
    except:
        x=0
    return x
```

```python
df['price'] = df['price'].apply(conversor)
```

```python
df
```

```python
items = df.groupby(['developer','year'])['id'].count().reset_index(name='cantItems')
items
```

```python
total_per_developer = df.groupby(['developer','year']).size().reset_index(name='items_per_year')
free_per_developer = df[df['price'] == 0].groupby(['year', 'developer']).size().reset_index(name='free_per_year')
```

```python
resultado = pd.merge(total_per_developer, free_per_developer, on=['developer','year'], how='left')
```

```python
resultado['free_per_year'] = resultado['free_per_year'].fillna(0)
```

```python
resultado['free_per_year_percent']=resultado['free_per_year']/resultado['items_per_year']*100
```

```python
resultado
```

```python
resultado.to_csv('developerResults.csv', index=False)
```
